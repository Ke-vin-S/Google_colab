```python
def analyze_csv(file_path, column_name):
    # Load CSV file
    df = pd.read_csv(file_path)

    # Extract the target column
    data = df[column_name].dropna()

    # Compute frequency
    frequency = data.value_counts()
    normalized_frequency = frequency / frequency.sum()
    cumulative_frequency = frequency.cumsum()
    normalized_cumulative_frequency = cumulative_frequency / cumulative_frequency.max()

    # Compute statistical measures
    mode = data.mode().tolist()
    median = data.median()
    mean = data.mean()
    data_range = data.max() - data.min()
    variance = data.var()
    std_dev = data.std()

    # Additional statistical measures
    min_value = data.min()
    q1 = data.quantile(0.25)
    q2 = median  # Median is Q2
    q3 = data.quantile(0.75)
    max_value = data.max()
    iqr = q3 - q1

    # Print results
    print("Attribute Type:", data.dtype)
    print("Frequency:\n", frequency)
    print("Normalized Frequency:\n", normalized_frequency)
    print("Cumulative Frequency:\n", cumulative_frequency)
    print("Normalized Cumulative Frequency:\n", normalized_cumulative_frequency)
    print("Mode:", mode)
    print("Median:", median)
    print("Mean:", mean)
    print("Range:", data_range)
    print("Variance:", variance)
    print("Standard Deviation:", std_dev)
    print("Min:", min_value)
    print("Lower Quartile (Q1):", q1)
    print("Median (Q2):", q2)
    print("Upper Quartile (Q3):", q3)
    print("Max:", max_value)
    print("Interquartile Range (IQR):", iqr)
```

```python
analyze_csv(data, "Trip Duration"	)
```
